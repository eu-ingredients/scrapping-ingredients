```python
import json
from pathlib import Path
import os

products_path = (
    Path(os.path.abspath("")) / "scrapping_ingredients" / "ah" / "products.json"
)
all_products = json.loads(products_path.read_bytes())
```

```python
all_products
```

```python
all_ingredients = []

for product in all_products.values():
    _ingredients = product.get("processed_ingredients", [])
    if _ingredients:
        all_ingredients += _ingredients

print(f"total ingredients: {len(all_ingredients)}")

unique_ingredients = set(all_ingredients)

print(f"unique raw ingredients {len(unique_ingredients)}")
```
